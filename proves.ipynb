```python
import re
def find_match(pattern, string):
    match = re.search(pattern, string)
    if match:
        return match.group()
    else:
        return 'no match.'
```

```python
pattern = re.compile(r'[^@]\b\w+\b')
s = '@united Had to explain to a very over eager flight attendant trying to ask about my vision that my @ccicanine was not a guide dog!'

find_match(pattern, s)
```
